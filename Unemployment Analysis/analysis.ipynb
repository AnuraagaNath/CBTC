## Data Cleaning

```python
df1['Date'] = pd.to_datetime(df1['Date'], format='%d-%m-%Y')
```

```python
df1.isna().sum()
```

```python
df1 = df1.dropna(axis=0)
```

```python
df1
```

```python
df2['Date'] = pd.to_datetime(df2['Date'], format='%d-%m-%Y')
```

```python
df2 = df2.drop(['Longitude', 'Latitude'], axis=1)
```

```python
df2 = df2.rename(columns={'Region.1':'Regional'})
```

```python
df2
```

```python
df1.query('Region=="Karnataka"')
```

```python
df2.query('Region=="Karnataka"')
```

There is a data discrepancy between the two data. To encounter this issue, the data from January 2020 in df1 is replaced by df2 data. 

```python
# removing all 2020 data from df1
df1 = df1[df1['Date'].dt.year != 2020]
```

```python
regions = df1['Region'].unique()

result = pd.DataFrame()

for region in regions:
    df1_region = df1[df1['Region']==region]
    df2_region = df2[df2['Region']==region]

    df1_region = pd.concat([df1_region, df2_region], ignore_index=True)

    result = pd.concat([result, df1_region], ignore_index=True)  
    
```

```python
result.query('Region=="Karnataka"')
```

```python
result['Regional'] = result['Regional'].bfill(axis=0)
```

```python
result['Area'] = result['Area'].ffill(axis=0)
```

```python
result[result['Region']=='Chandigarh']
```

```python
chandigarh_index = result[result['Region']=='Chandigarh'].index
for idx in chandigarh_index:
    result.at[idx, 'Regional'] = 'North'
```

```python
result[result['Region']=='Chandigarh']
```

```python
result.isna().sum()
```

```python
result = result.drop(['Frequency'], axis=1)
```

```python
result
```

```python
result['Date'] = pd.to_datetime(result['Date']).dt.date
```

```python
# result.to_parquet('unemployment_rate.parquet')
```

```python
def getRegionWisePlots(y):
    fig, axs = plt.subplots(7, 4, figsize=(15, 20))  
    regions = result['Region'].unique()
    
    for i, region in enumerate(regions):
        row = i // 4
        col = i % 4
        ax = axs[row, col]
        
        region_data = result[result['Region'] == region]
        
        for label in ax.get_xticklabels():
            label.set_rotation(45) 
            
        sns.lineplot(data=region_data, x='Date', y=y, ax=ax, errorbar=None)
        ax.set_title(region)
    
    plt.tight_layout()
    return plt.show()
```
